```python
#line width
#this gives the line width of each line
plt.plot(batsman['index'],batsman['V Kohli'], color= 'green', linestyle='dashed', linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle= 'dotted', linewidth= 2)
plt.title('Rohit Sahrms vs Virat Kolhi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#manuplating the markers
#markers can be D, ., +, <, >, O, 
plt.plot(batsman['index'],batsman['V Kohli'], color= 'green', linestyle='dashed', linewidth=3, marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle= 'dotted', linewidth= 2, marker='+')
plt.title('Rohit Sahrms vs Virat Kolhi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#controlling the markers size
#markers size can be any number. bigger number means big size. 
plt.plot(batsman['index'],batsman['V Kohli'], color= 'green', linestyle='dashed', linewidth=3, marker='D', markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle= 'dotted', linewidth= 2, marker='+', markersize= 20)
plt.title('Rohit Sahrms vs Virat Kolhi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#legend ----> location
#to refer which line is for which thing we called it labelling. 
#labelling is done through legends
plt.plot(batsman['index'],batsman['V Kohli'], color= 'green', linestyle='dashed', linewidth=3, marker='D', markersize=10, label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle= 'dotted', linewidth= 2, marker='+', markersize= 20,label='Rohit')
plt.title('Rohit Sahrms vs Virat Kolhi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

##here you do not see the name of the labels because there is one more important thing that is legend, which needs to be passed.
##if we do not pass the legend then even though we put the label, it will not work.
```

```python
#here we are going to put the legend so that we can see how the upper one works after putitng legend.
plt.plot(batsman['index'],batsman['V Kohli'], color= 'green', linestyle='dashed', linewidth=3, marker='D', markersize=10, label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle= 'dotted', linewidth= 2, marker='+', markersize= 20,label='Rohit')
plt.title('Rohit Sahrms vs Virat Kolhi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

```python
#in above you may ask why the legend is on the top left cornor, can we make it be situated other place.
#the reason is the defult positon is plt.legend(loc='best')
#we can make changes as following plt.legend(loc='left, right, lower left, lower right, center left, center right, lower center, upper center, upper right, lower right')
plt.plot(batsman['index'],batsman['V Kohli'], color= 'green', linestyle='dashed', linewidth=3, marker='D', markersize=10, label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'], color='black',linestyle= 'dotted', linewidth= 2, marker='+', markersize= 20,label='Rohit')
plt.title('Rohit Sahrms vs Virat Kolhi Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='upper center')
```

```python
#problem with categories overlapping when there are long name
children=[10,20,40,10,30]
colors=['red red red red', 'blue blue blue blue','green green green green','yellow yellow yellow yellow', 'pink pink pink pink']
plt.bar(colors,children)
```

```python
#solution for problem with categories overlapping
#here we gave 70 degree rotation in xticks
children=[10,20,40,10,30]
colors=['red red red red', 'blue blue blue blue','green green green green','yellow yellow yellow yellow', 'pink pink pink pink']
plt.bar(colors,children)
plt.xticks(rotation=70) # Rotating x-axis labels to 70 degree for better visibility
```

```python
#solution for problem with categories overlapping
#here we used vertical
children=[10,20,40,10,30]
colors=['red red red red', 'blue blue blue blue','green green green green','yellow yellow yellow yellow', 'pink pink pink pink']
plt.bar(colors,children)
plt.xticks(rotation='vertical') # Rotating x-axis labels vertically for better visibility
```
